```python
# Here we take rf = 0.02 as default 
# and we append 0.002 in some month as the model shows all the index would get down and we would 
# invest on risk-free assets
exp_ret = []
vol = []
sr = []
act_ret = []


for i in range(4):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)

act_ret.append(0.002)



for i in range(5,6):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)

act_ret.append(0.002)

for i in range(7,10):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)

act_ret.append(0.002)

for i in range(11,18):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)
    
act_ret.append(0.002)

for i in range(19,24):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)
    
act_ret.append(0.002)

for i in range(26,29):
    df = pd.DataFrame()
    for ind in ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']:
        dt = pd.read_csv('%s.csv'%ind,index_col=0)
        df = pd.concat([df,dt.iloc[:,i]],axis = 1)
    df.columns = ['RU20INTR','RUMCINTR','LBUSTRUU','NDDUWXUS','LG30TRUU','BCOMTR']
    
    
    mu = expected_returns.mean_historical_return(df)
    S = risk_models.sample_cov(df)
        
    # Optimise for maximal Sharpe ratio
    ef = EfficientFrontier(mu, S)
    raw_weights = ef.max_sharpe()
    cleaned_weights = ef.clean_weights()
        
    exp_ret.append(ef.portfolio_performance(verbose=False)[0])
    vol.append(ef.portfolio_performance(verbose=False)[1])
    sr.append(ef.portfolio_performance(verbose=False)[2])
        
    wt = list(cleaned_weights.values())
    start_val = sum(wt*df_.iloc[i*30])
    end_val = sum(wt*df_.iloc[(i+1)*30])
    rrr = end_val/start_val - 1 
    act_ret.append(rrr)
```
